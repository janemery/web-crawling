#  Atividade Prática 03 –  Web Crawling & Scraping com Scrapy  
## Módulo: Web Crawling, Scraping e Coleta de Dados Automatizada

## Contexto

Nesta atividade, você será responsável por coletar informações automaticamente da web utilizando **Scrapy**.  
Seu objetivo é construir um *crawler* capaz de navegar por **ao menos 10 páginas**, extrair informações estruturadas e produzir um conjunto de dados para análise.

Esse tipo de automação é comum em cenários como:

- Pesquisa de mercado  
- Monitoramento de preços  
- Coleta de dados para análises estatísticas  
- Investigação acadêmica  

Depois de coletar os dados, você irá importá-los neste notebook e realizar uma **análise de estatística descritiva**.


## Instruções

- Criar e configurar um projeto Scrapy.
- Criar um spider com paginação (mínimo **10 páginas**).
- Escolha um site público com **paginação**, como:
    - Livros  
    - Produtos  
    - Filmes  
    - Citações  
    - Cursos  
    - Vagas de emprego 
- Extrair pelo menos **3 campos** por item.
- Utilizar o **pipeline** para tratar algum campo
- Exportar os dados como CSV ou JSON.
- Calcular estatísticas descritivas básicas.
- Criar ao menos 1 gráfico simples.

## Entrega

Arquivo zip com a pasta do scrapy e o notebook com a análise descritiva.
