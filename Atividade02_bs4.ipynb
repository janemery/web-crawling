#  Atividade Prática 02 –  Web Crawling & Scraping com BeautifulSoup  
## Módulo: Web Crawling, Scraping e Coleta de Dados Automatizada

## Objetivos da Atividade

- Entender a estrutura HTML de uma página real.  
- Utilizar `requests` e `BeautifulSoup` para extrair dados.  
- Coletar informações relevantes (ex.: título, autor, data, categoria).  
- Estruturar tudo em um DataFrame.  
- Fazer uma análise estatística descritiva simples sobre os dados coletados.  
- Interpretar os resultados obtidos.

---

## Contexto da Atividade

Nesta atividade, você irá praticar os fundamentos de **extração de dados da web** utilizando a biblioteca **BeautifulSoup**.  
O objetivo é simular um cenário real: **coletar informações de uma página de notícias**, organizar esses dados e realizar uma **análise estatística descritiva simples**.

Em aplicações reais, profissionais de dados frequentemente coletam manchetes, datas de publicação, autores, categorias e outros atributos para entender tendências, frequência de temas e padrões temporais.

---

## Instruções

1. Escolha uma **página de notícias** (ex.: G1, UOL, Folha, CNN Brasil, TecMundo, etc.).  
2. Inspecione o HTML e identifique os elementos onde estão as manchetes.  
3. Use `requests` para baixar o HTML.  
4. Use `BeautifulSoup` para extrair:
   - Título da notícia  
   - Data (se existir)  
   - Categoria (se existir)  
   - Link da notícia  
5. Organize os dados em um DataFrame pandas.  
6. Realize **pelo menos uma análise descritiva**, como:
   - Contagem de categorias  
   - Frequência de palavras nos títulos  
   - Tamanho médio dos títulos  
   - Quantidade de notícias por data  
7. Exiba gráficos simples (opcional).
