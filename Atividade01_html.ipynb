#  Atividade Prática 01 – Construção de Página HTML  
## Módulo: Web Crawling, Scraping e Coleta de Dados Automatizada

## Objetivo da Atividade

O objetivo desta atividade é reforçar os conhecimentos de **HTML** essenciais para Web Scraping. 

## Contexto

Antes de começarmos a extrair dados de páginas reais, é importante entender como essas páginas são estruturadas.  
HTML é o alicerce da web, e um bom scraper precisa reconhecer:

- tags,
- hierarquias,
- atributos,
- classes e IDs,
- relações entre elementos,
- listas, tabelas, seções, links, imagens e muito mais.

---

## Instruções da Atividade

Crie um arquivo **`teste.html`** contendo:

### ✔ Estrutura básica obrigatória:
- `<!DOCTYPE html>`
- `<html>`, `<head>`, `<body>`
- `<title>` com algum nome criativo

---

### ✔ Dentro do `<body>`, utilize **o máximo de tags HTML possíveis**, incluindo mas não limitado a:

#### **Textos e estruturação**
- `<h1>` até `<h6>`
- `<p>`
- `<span>`
- `<div>`
- `<section>`  
- `<article>`  
- `<header>`  
- `<footer>`  

#### **Formatação de texto**
- `<b>`, `<strong>`
- `<i>`, `<em>`
- `<mark>`
- `<small>`
- `<blockquote>`
- `<pre>`
- `<code>`

#### **Listas**
- `<ul>` e `<ol>`
- `<li>`
- `<dl>`, `<dt>`, `<dd>`

#### **Tabelas**
- `<table>`
- `<tr>`, `<td>`, `<th>`
- `<thead>`, `<tbody>`

#### **Links e mídias**
- `<a>`
- `<img>`

#### **Formulários**
- `<form>`
- `<input>` (use pelo menos 3 tipos diferentes)
- `<select>` + `<option>`
- `<textarea>`
- `<button>`

#### **Outras tags interessantes**
- `<nav>`
- `<aside>`
- `<figure>` + `<figcaption>`
- `<hr>`
- `<br>`
- Comentários `<!-- -->`

---

## Regras da Atividade

1. Use **ao menos 15 tags HTML diferentes**.  
2. Sugestão, a página deve parecer uma **mini matéria/jornal**, como se fosse uma notícia ou artigo simples.  
3. Inclua elementos que facilitam posteriormente a extração de dados via scraping.  
   - Classes  
   - IDs  
   - Atributos personalizados (`data-*`)  
4. Mantenha uma estrutura organizada e legível.

---

### Pontos Bônus

Use `CSS` para deixar sua página ainda mais interessante.


## Entrega

Envie o arquivo `teste.html` no AVA.
